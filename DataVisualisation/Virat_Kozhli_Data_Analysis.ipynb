Q3:Consider the following statements and choose the correct option

     I - Virat has played the maximum number of matches in 2011
     II - Virat has the highest run average in the year 2017
     III - Virat has the maximum score in a single match and the highest run average in the year 2016.

Which of the above statements is/are false?

    a)I and II
    b)I and III
    c)II
    d)III



```python
df
```

```python
#Fetch 'Year' from 'Start Date' Convert the 'Start date' column into datetime datatype
df['Start Date'] = df['Start Date'].apply(pd.to_datetime) 
df.info()
df['Year'] = df['Start Date'].dt.to_period('Y')
df
```

```python
#I - Virat has played the maximum number of matches in 2011
df.groupby(['Year'])['Year'].count()
```

```python
#II - Virat has the highest run average in the year 2017
df.groupby(['Year'])['Runs'].sum() / df.groupby(['Year'])['Year'].count()
```

```python
#III - Virat has the maximum score in a single match and the highest run average in the year 2016.
a = df.Runs.max()
df[df["Runs"] == a]
```
