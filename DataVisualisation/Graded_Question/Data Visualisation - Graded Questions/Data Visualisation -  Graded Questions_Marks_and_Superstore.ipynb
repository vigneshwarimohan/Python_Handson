Question 2:
Data Visualisation
You can observe that the column Ship Mode has 4 categories - First Class, Second Class, Same Day and Standard Class, and the Segment has 3 - Consumer, Home Office, Corporate.

Now let's say you want to visualise how the average Profit varies across every Segment - Ship Mode combination (like Consumer - First Class, Corporate - Standard Class and so on).
Which of the following visualisations can be utilised for the same?

Here, you're trying to analyse how a numerical variable changes across two categorical variables. So, you can create a table where the rows denote the Segment variable whereas the columns denote the Ship Mode. Thus, each cell would correspond to a particular Segment - Ship Mode pair and its value would be given by the corresponding average Profit for the same. Once this grid is created, you can easily use a heat map to plot the values 
and visualise how the average Profit changes across the various Segment - Ship Mode combinations.

```python
r = df3[["Ship Mode","Segment","Profit"]]
r
```

```python
res = pd.pivot_table(data = r,index = 'Ship Mode',columns = 'Segment',values = 'Profit',aggfunc = np.sum)
res
```

```python
sns.heatmap(res,cmap = "Greens", annot=True)
plt.show()
```
